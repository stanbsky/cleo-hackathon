```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import seaborn as sns
import scipy
```

```python
import json
```

```python
with open('c1_transactions.json') as json_file:
    trans = json.load(json_file)
```

```python
merged = []
for _, v in trans.items():
    merged.extend(v)
print(merged)
```

```python
trans = pd.DataFrame(merged)
trans.columns
```

```python
trans.head()
```

```python
trans = trans.drop(columns=['account_owner','authorized_date','authorized_datetime','datetime','location',
                    'iso_currency_code','payment_channel','payment_meta','pending','pending_transaction_id',
                   'transaction_code','transaction_type','unofficial_currency_code'])
```

```python
trans.category = pd.Series(['|'.join(x) for x in trans.category])
```

```python
len(trans.category.unique())
```

```python
transfers = ['Transfer|Payroll','Transfer|Withdrawal','Transfer|Debit','Transfer|Third Party|PayPal']
trans[trans.category.isin(transfers)]
```

```python
trans.date = pd.to_datetime(trans.date, format="%Y-%m-%d")
```
