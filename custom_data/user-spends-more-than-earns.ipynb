```python
from datetime import date
from dateutil.relativedelta import relativedelta

start = date(2020,1,1)
spend = debit + credit
print("CC payment, interest, CC balance")
for i in range(12):
    rng = np.random.default_rng()
    debits = rng.integers(low=(debit*0.85), high=(debit*1.15), size=1)[0]
    credits = rng.integers(low=(credit*0.85), high=(credit*1.15), size=1)[0]
    interest = bal * 0.15 / 12
    bal -= cc_payment
    bal += interest
    bal += credits
    month = start.strftime('%Y-%m-%d')
    t = t.append(
        pd.DataFrame(
            [
                [month, income*-1, 'credit', 'paycheck'],
                [month, debits, 'debit', 'debit_card'],
                [month, cc_payment, 'debit', 'cc_payment'],
                [month, credits, 'debit', 'credit_card']
            ], columns=cols
        )
    )
    print(cc_payment, interest, bal)
    start = start + relativedelta(months=+1)
t
```
